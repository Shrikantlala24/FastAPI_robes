```python
from pydantic import BaseModel, Field
```

```python
class Patient(BaseModel):
    name : str
    age : int = Field(gt = 0)
```

```python
# help(Field)
```

```python
def insert(p : Patient) :
    print(p.name)
    print(p.age)
```

```python
p1 = Patient(**{
    'name':'Shri',
    'age':30
})
```

```python
insert(p1)
```
